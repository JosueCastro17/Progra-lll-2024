Calcular area

```python
def calcular_area(de, a, cantidad):
   valores = [1, 10000, 7000, 2722500, 302500, 16]
   return valores[a] / valores[de] * cantidad

def menu_area():
   print("Bienvenidos a Conversor Area")
   print("0. Salir")
   print("1. Manzana")
   print("2. Vara Cuadrada")
   print("3. Metro Cuadrado")
   print("4. Pie Cuadrado")
   print("5. Yarda Cuadrada")
   print("6. Tareas")


   de = int(input("De: ")) -1

   print("")
   print("1. Manzana")
   print("2. Vara Cuadrada")
   print("3. Metro Cuadrado")
   print("4. Pie Cuadrado")
   print("5. Yarda Cuadrada")
   print("6. Tareas")

   a = int(input("A: "))-1

   cantidad = float(input("Cantidad: "))

   resultado = calcular_area(de, a, cantidad)

   print("Resultado: ", resultado)

   return f"El resultado de la conversión es: {resultado}"


def menu_principal():
   print("1. Area")
   print("0. Salir")

   opcion = int(input("Opcion: "))

   while True:
    if opcion == 1:
      print(menu_area())
      break

menu_principal()
```

```python
def cajero_automatico(cantidad):
  if cantidad <= 0:
    return "Ingrese una cantidad válida."

  billetes = [20, 10, 5, 1]
  monedas = [0.25, 0.10, 0.05, 0.01]

  resultado = ""

  for billete in billetes:
    num_billetes = int(cantidad // billete)
    if num_billetes > 0:
      resultado += f"{num_billetes} billete(s) de a {billete}\n"
      cantidad -= num_billetes * billete

  for moneda in monedas:
    num_monedas = int(cantidad // moneda)
    if num_monedas > 0:
      resultado += f"{num_monedas} moneda(s) de {moneda}\n"
      cantidad -= num_monedas * moneda

  return resultado

cantidad = float(input("Ingrese la cantidad a retirar: "))
print(cajero_automatico(cantidad))
```
