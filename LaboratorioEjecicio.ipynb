```python
def Monedas():
    print("Bienvenido a conversor Monedas")

    monedas = ["Dolar", "Quetzal", "Colon Salvadoreno", "Euro", "Peso Mexicano", "Lempira", "Cordoba", "Peso Colombiano", "Peso Argentino", "Peso Chileno"]
    tasas = [1.0, 7.75, 8.75, 0.92, 18.83, 24.77, 36.82, 4071.00, 936.18, 932.33]

    while True:
        for i, moneda in enumerate(monedas):
            print(f"{i + 1}. {moneda}")

        origen = int(input("Moneda de origen: ")) - 1
        destino = int(input("Moneda de destino: ")) - 1

        if 0 <= origen < len(monedas) and 0 <= destino < len(monedas):
            cantidad = float(input("Cantidad: "))
            print(f"{cantidad} {monedas[origen]} = {(cantidad / tasas[origen]) * tasas[destino]:.2f} {monedas[destino]}")
        else:
            print("Selección de moneda inválida, intenta nuevamente.")

        if input("¿Otra conversión? (s/n): ").lower() != 's':
            break

def Longitud():
    print("Bienvenido a conversor de Longitud")

    longitud = ["Metros", "Kilometro", "Hectometro", "Decametro", "Decimetro", "Centimetros", "Milimetro", "Pulgada", "Pie", "Yarda"]
    tasas = [1.0, 0.001, 0.01, 0.1, 10.0, 100.0, 1000.0, 39.37, 3.28, 1.09]

    while True:
        for i, longitudes in enumerate(longitud):
            print(f"{i + 1}. {longitudes}")

        origen = int(input("Moneda de origen: ")) - 1
        destino = int(input("Moneda de destino: ")) - 1

        if 0 <= origen < len(longitud) and 0 <= destino < len(longitud):
            cantidad = float(input("Cantidad: "))
            print(f"{cantidad} {longitud[origen]} = {(cantidad / tasas[origen]) * tasas[destino]:.2f} {longitud[destino]}")
        else:
            print("Selección de moneda inválida, intenta nuevamente.")

        if input("¿Otra conversión? (s/n): ").lower() != 's':
            break

def Masa():
    print("Bienvenido a conversor de Masa")

    masas = ["Libra", "Gramo", "miligramo", "microgramo", "tonelada larga", "onza", "Stone", "kilogramo", "Quintal", "Tonelada Corta"]
    tasas = [1, 453.59, 453592.37, 453592370, 0.0004464, 16.0, 0.07142857, 0.4535, 0.01, 0.0005 ]

    while True:
        for i, masa in enumerate(masas):
            print(f"{i + 1}. {masa}")

        origen = int(input("Moneda de origen: ")) - 1
        destino = int(input("Moneda de destino: ")) - 1

        if 0 <= origen < len(masas) and 0 <= destino < len(masas):
            cantidad = float(input("Cantidad: "))
            print(f"{cantidad} {masas[origen]} = {(cantidad / tasas[origen]) * tasas[destino]:.1f} {masas[destino]}")
        else:
            print("Selección de moneda inválida, intenta nuevamente.")

        if input("¿Otra conversión? (s/n): ").lower() != 's':
            break

def Almacenamiento():
    print("Bienvenido al conversor de Almacenamiento")

    unidades = ["Megabyte", "Bit", "Kilobyte", "Byte", "Gigabyte", "Terabyte", "Petabyte", "Exabyte", "Zettabyte", "Yottabyte"]

    tasas = {
        "Bit": 1,
        "Byte": 8,
        "Kilobyte": 8 * 1000,
        "Megabyte": 8 * 1000 * 1000,
        "Gigabyte": 8 * 1000 * 1000 * 1000,
        "Terabyte": 8 * 1000 * 1000 * 1000 * 1000,
        "Petabyte": 8 * 1000 * 1000 * 1000 * 1000 * 1000,
        "Exabyte": 8 * 1000 * 1000 * 1000 * 1000 * 1000 * 1000,
        "Zettabyte": 8 * 1000 * 1000 * 1000 * 1000 * 1000 * 1000 * 1000,
        "Yottabyte": 8 * 1000 * 1000 * 1000 * 1000 * 1000 * 1000 * 1000 * 1000
    }

    while True:
        for i, unidad in enumerate(unidades):
            print(f"{i + 1}. {unidad}")

        origen = int(input("Unidad de origen (número): ")) - 1
        destino = int(input("Unidad de destino (número): ")) - 1

        if 0 <= origen < len(unidades) and 0 <= destino < len(unidades):
            cantidad = float(input("Cantidad: "))
            unidad_origen = unidades[origen]
            unidad_destino = unidades[destino]
            tasa_origen = tasas[unidad_origen]
            tasa_destino = tasas[unidad_destino]
            resultado = (cantidad * tasa_origen) / tasa_destino
            print(f"{cantidad} {unidad_origen} = {resultado:.15f} {unidad_destino}")
        else:
            print("Selección de unidad inválida, intenta nuevamente.")

        if input("¿Otra conversión? (s/n): ").lower() != 's':
            break

def Tiempo():
    print("Bienvenido al conversor de Tiempo")

    Tiempos = ["Minuto", "Segundos", "Milisegundos", "Microsegundos", "Horas", "Días", "Semanas", "Meses", "Años", "Siglos"]
    tasas = [60, 1, 1e-3, 1e-6, 3600, 86400, 604800, 2629744, 31557600, 3155760000]

    while True:
        for i, tiempo in enumerate(Tiempos):
            print(f"{i + 1}. {tiempo}")

        origen = int(input("Unidad de origen (número): ")) - 1
        destino = int(input("Unidad de destino (número): ")) - 1

        if 0 <= origen < len(Tiempos) and 0 <= destino < len(Tiempos):
            cantidad = float(input("Cantidad: "))
            tasa_origen = tasas[origen]
            tasa_destino = tasas[destino]
            resultado = (cantidad * tasa_origen) / tasa_destino
            print(f"{cantidad} {Tiempos[origen]} = {resultado:.6f} {Tiempos[destino]}")
        else:
            print("Selección de unidad inválida, intenta nuevamente.")

        if input("¿Otra conversión? (s/n): ").lower() != 's':
            break


def Volumen():
    print("Bienvenido al conversor de Volumen")

    Unidades = ["Litros", "Mililitros", "Centímetros cúbicos", "Decímetros cúbicos", "Metros cúbicos", "Galones (EE.UU.)", "Tazas (EE.UU.)", "Cucharadas (EE.UU.)", "Galones (Imperiales)"]
    tasas = [1, 1e3, 1e3, 1e1, 1e-3, 0.264172, 4.22675, 67.628, 0.219969]

    while True:
        print("\nUnidades disponibles:")
        for i, unidad in enumerate(Unidades):
            print(f"{i + 1}. {unidad}")

        origen = int(input("Unidad de origen (número): ")) - 1
        destino = int(input("Unidad de destino (número): ")) - 1

        if 0 <= origen < len(Unidades) and 0 <= destino < len(Unidades):
            cantidad = float(input("Cantidad: "))
            cantidad_en_litros = cantidad / tasas[origen]
            resultado = cantidad_en_litros * tasas[destino]
            print(f"{cantidad} {Unidades[origen]} = {resultado:.2f} {Unidades[destino]}")
        else:
            print("Selección de unidad inválida, intenta nuevamente.")

        if input("¿Otra conversión? (s/n): ").lower() != 's':
            break

def Area():
    print("Bienvenido al conversor de Área")

    Unidades = ["Hectárea", "Metro cuadrado", "Kilómetro cuadrado", "Manzana", "Milímetro cuadrado", "Tarea", "Yarda cuadrada", "Pie cuadrado", "Pulgada cuadrada", "Milla cuadrada"]

    tasas = [
        1e4,
        1,
        1e6,
        7000,
        1e-6,
        625,
        0.836127,
        0.092903,
        6.4516e-4,
        2.58999e6
    ]

    while True:
        print("Unidades disponibles:")
        for i, unidad in enumerate(Unidades):
            print(f"{i + 1}. {unidad}")

        origen = int(input("Unidad de origen (número): ")) - 1
        destino = int(input("Unidad de destino (número): ")) - 1

        if 0 <= origen < len(Unidades) and 0 <= destino < len(Unidades):
            cantidad = float(input("Cantidad: "))
            cantidad_en_metros_cuadrados = cantidad * tasas[origen]
            resultado = cantidad_en_metros_cuadrados / tasas[destino]
            print(f"{cantidad} {Unidades[origen]} = {resultado:.2f} {Unidades[destino]}")
        else:
            print("Selección de unidad inválida, intenta nuevamente.")

        if input("¿Otra conversión? (s/n): ").lower() != 's':
            break

def menuPrincipal():
    print("Bienvenido al Menu Principal de Conversores")
    print("1. Monedas")
    print("2. Longitud")
    print("3. Masa")
    print("4. Almacenamiento")
    print("5. Tiempo")
    print("6. Volumen")
    print("7. Area")
    print("8. Salir ")

while True:
    menuPrincipal()
    opcion_Menu = input("Seleccione el conversor del 1 al 6: ")

    if opcion_Menu == '1':
        Monedas()
    elif opcion_Menu == '2':
        Longitud()
    elif opcion_Menu == '3':
        Masa()
    elif opcion_Menu == '4':
        Almacenamiento()
    elif opcion_Menu == '5':
        Tiempo()
    elif opcion_Menu == '6':
        Volumen()
    elif opcion_Menu == '7':
        Area()
    elif opcion_Menu == '8':
        print("Saliendo del programa...")
        break
    else:
        print("Opción inválida")
```

```python
#DEDUCCIONES DE SUELDO

def calcular_deducciones(sueldo):
    isss = sueldo * 0.03
    afp = sueldo * 0.0625
    restarAFP = sueldo - afp
    restar_isss = sueldo - (afp + isss)
    isss2 = 0

    if restarAFP >= 0.01 and restarAFP <= 487.60:
        isr = 0
    elif restarAFP >= 487.61 and restarAFP <= 642.85:
        isr = (restarAFP - 487.60) * 0.1 + 17.48
    elif restarAFP >= 642.86 and restarAFP <= 915.81:
        isr = (restarAFP - 642.85) * 0.1 + 32.71
    elif restarAFP >= 915.82 and restarAFP <= 2058.67:
        isr = (restarAFP - 915.81) * 0.2 + 60
    elif restarAFP >= 2058.68:
        isr = (restarAFP - 2058.67) * 0.3 + 288.57

    if restar_isss >= 0.01 and restar_isss <= 472.01:
        isss2 = 0
    elif restar_isss >= 472.01 and restar_isss <= 895.24:
        isss2 = (restar_isss - 472.01) * 0.1 + 17.67
    elif restar_isss >= 895.25 and restar_isss <= 2038.10:
        isss2 = (restar_isss - 895.24) * 0.2 + 60
    elif restar_isss >= 2038.11:
        isss2 = (restar_isss - 2038.10) * 0.3 + 288.57

    total_deducciones = isss + afp + isr + isss2
    sueldo_neto = sueldo - total_deducciones

    return {
        'ISSS': isss,
        'AFP': afp,
        'ISR': isr,
        'Bienestar': isss2,
        'Total Deducciones': total_deducciones,
        'Sueldo Neto': sueldo_neto
    }

sueldo = float(input("Ingresa tu sueldo: "))

resultados = calcular_deducciones(sueldo)

print(f"ISSS: {resultados['ISSS']:.2f}")
print(f"AFP: {resultados['AFP']:.2f}")
print(f"ISR: {resultados['ISR']:.2f}")
print(f"Bienestar: {resultados['Bienestar']:.2f}")
print(f"Total Deducciones: {resultados['Total Deducciones']:.2f}")
print(f"Sueldo Neto: {resultados['Sueldo Neto']:.2f}")
```

```python
from fractions import Fraction

def unidades_a_cajas(unidades):
    cajas_completas = unidades // 25
    unidades_restantes = unidades % 25
    if unidades_restantes == 0:
        return f"{cajas_completas} cajas"
    else:
        fraccion = Fraction(unidades_restantes, 25)
        return f"{cajas_completas} cajas y {unidades_restantes} unidades"

def cajas_a_unidades(cajas):
    unidades = cajas * 25
    return f"{cajas} cajas equivalen a {unidades} unidades."

while True:
    print("Conversor de Cajas y Unidades")
    print("1. Convertir unidades a cajas")
    print("2. Convertir cajas a unidades")
    print("3. Salir")

    opcion = int(input("Seleccione una opción: "))

    if opcion == 1:
        unidades = int(input("Ingrese la cantidad de unidades: "))
        resultado = unidades_a_cajas(unidades)
        print(f"{unidades} unidades equivalen a {resultado}.")
    elif opcion == 2:
        cajas = float(input("Ingrese la cantidad de cajas: "))
        resultado = cajas_a_unidades(cajas)
        print(resultado)
    elif opcion == 3:
        break
    else:
        print("Opción inválida, seleccione una opción válida.")
```

```python
#Varianza de una serie de números:
from math import sqrt

class estadistica:
    def media(self, serie):
        suma = 0
        for num in serie:
            suma += num
        return suma / len(serie)

    def tipica(self, serie):
        media = self.media(serie)
        suma = 0
        for num in serie:
            suma += (num - media) ** 2
        return sqrt(suma / len(serie))

    def varianza(self, serie):
        media = self.media(serie)
        suma = 0
        for num in serie:
            suma += (num - media) ** 2
        return suma / len(serie)

n = int(input("Cuantos numeros va a ingresar: "))
serie = []
for num in range(1, n + 1):
    serie.append(int(input("Num: ")))

objEstadistica = estadistica()
print("La media es: ", objEstadistica.media(serie))
print("La desviacion tipica es: ", round(objEstadistica.tipica(serie), 3))
print("La varianza es: ", round(objEstadistica.varianza(serie), 3))
```

```python
# Frecuencia de cada número:
from collections import Counter

numeros = list(map(float, input("Introduce una serie de números separados por espacio: ").split()))
frecuencia = Counter(numeros)

print("Frecuencia de los números introducidos:")
for num, freq in frecuencia.items():
    print(f"Número: {num}, Frecuencia: {freq}")
```
